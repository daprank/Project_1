### 2 Первичная очистка данных.

### 2.1 Заполнение пропусков

```python
def categorial_fillna(df, column_with_nan, categorial_column, subcategorial_column, method_fill):
    categories = df[categorial_column].unique()
    subcategories = df[subcategorial_column].unique()
    column_without_nan = column_with_nan
    for category in categories:
        for subcategory in subcategories:
            if method_fill == 'median':
                df.loc[(df[categorial_column]==category)&(df[subcategorial_column]==subcategory), column_with_nan] = df[(df[categorial_column]==category)&(df[subcategorial_column]==subcategory)][column_with_nan].fillna(value = df[df[categorial_column]==category][column_with_nan].sort_values().median())
            elif method_fill == 'mean':
                df.loc[(df[categorial_column]==category)&(df[subcategorial_column]==subcategory), column_with_nan] = df[(df[categorial_column]==category)&(df[subcategorial_column]==subcategory)][column_with_nan].fillna(value = df[df[categorial_column]==category][column_with_nan].sort_values().mean())
                
    return df
initial_df['days_employed'] = initial_df['days_employed'].abs()
new_df = categorial_fillna(initial_df, 'days_employed', 'dob_years', 'education_id', 'mean')
new_df = categorial_fillna(initial_df, 'total_income', 'dob_years', 'income_type', 'median')
new_df = new_df.loc[new_df['gender']!='XNA'].reset_index(drop = True)
new_df.info()
```

```python
new_df['children'] = new_df['children'].replace(20, 2)
new_df['children'] = new_df['children'].replace(-1, 1)
print(new_df.loc[new_df['children']==-1])
```

### 2.3 Результаты первичной очистки

* Пропуски в 'days_employed' заменены средними значениями (метод .mean()) этого поля из соответствующих категорий 'dob_years' и 'education_id'. Категоризация по данным признакам была учтена, поскольку они больше остальных связаны со стажем работы человека. Непосредственно для замены использовался метод .fillna(). Также, предварительно перед заменой все значения были взяты по модулю, поскольку большое количество отрицательных значений сказывалось на точности среднего (а отрицательных быть не может).
* Пропуски в 'total_income' заменены медианными значениями (метод .median()) этого поля из соответствующих категорий 'dob_years' и 'income_type'. Категоризация по данным признакам была учтена, поскольку они больше остальных связаны с уровнем заработка человека. Непосредственно для замены использовался метод .fillna().
* Удалена строка, содержащая в поле 'gender' значение 'XNA'. 
* Значения 20 и -1 в 'children' заменены на 2 и 1 соответственно.

### 3.3 Перевод строковых значений в набор лемм

```python
from pymystem3 import Mystem
m = Mystem()
def lemmatization(row):
    lemma = ''.join(m.lemmatize(row))
    return lemma
new_df['purpose_lemmatize'] = new_df['purpose'].apply(lemmatization)
new_df.info()
```

### Вывод

Проведена лемматизация данных столбца 'purpose'. К каждой строке применена функция lemmatization, которая на вход получает строку, разбивает ее содержимое на леммы, объединяет строку методом .join() и возвращает целую строку - набор лемм исходной строки.

### 3.4 Формирование категоризирующих полей для дальнейшего анализа категорий

```python
MIN = int(new_df['total_income'].min())
MAX = int(new_df['total_income'].max())
def income_category(row):
    if row < 45000:
        return 'Уязвимое население'
    elif row < 75000:
        return 'Низший средний класс'
    elif row < 130000:
        return 'Средний средний класс'
    elif row < 600000:
        return 'Высший средний класс'
    else:
        return 'Богатые'
new_df['income_category'] = new_df['total_income'].apply(income_category)
purpose_categories = ['жилье', 'автомобиль', 'образование', 'свадьба']
def category(row):
    if 'коммерческий' in row:
        return 'коммерческая недвижимость'
    for element in purpose_categories:
        if element in row:
            return element
    if 'недвижимость' in row:
        return 'жилье'
new_df['purpose_categories'] = new_df['purpose_lemmatize'].apply(category)
print(new_df['purpose_categories'].unique())
```

### Вывод

* Данные категоризированы до достатку. Люди с достатком менее 45 тысяч отнесены к уязвимому населению. С достатком от 45 до 75 тысяч к низшему среднему классу. Люди, зарабатывающие от 75 до 130 тысяч признаны средним средним классом. Тем, чей достаток находится в промежутке от 130 до 600 тысяч, присвоена группа "Высший средний класс". Остальные признаны богатыми.
* Даные категоризированы по цели займа. Определены категории: жилье, автомобиль, образование, свадьба.

### 4.2 Сравнительный анализ категорий по семейному положению в отношении наличия задолженности

```python
df_grouped = new_df.groupby('family_status')['debt'].agg(['sum', 'count', 'mean'])
print(df_grouped.sort_values('mean'))
def married(row):
    if (row == 'вдовец / вдова') or (row == 'в разводе') or (row == 'женат / замужем'):
        return 'Брак был или есть'
    else:
        return 'Никогда не состоял(а) в браке'
new_df['married'] = new_df['family_status'].apply(married)
df_grouped = new_df.groupby('married')['debt'].agg(['sum', 'count', 'mean'])
print(df_grouped.sort_values('mean'))
```

### Вывод

Из данных очевидно, что самые недобропорядочные заемщики - одинокие люди, не бывшие в браке. Интересно, что самые добропорядочные заемщики это те, кто потерял супруга или находится в разводе (6,5% и 7,1% имеют задолженность). Кроме того, мы можем сказать, что люди, предпочитающие не регистрировать официальный брак, на 1,8% чаще имеют задолженность по кредиту, чем официально состоящие в браке. Можно сделать предположение, что официальная регистрация и кредитная добропорядочность коррелируют с фактором "ответственности" людей.
<br>
Между теми, кто был или находится в официальном браке не определяется существенная разница по доле людей с задолженностью (менее 1% в любом случае). Так, данные можно поделить по признаку наличия в жизни официального брака. Гипотеза: люди, которые никогда не состояли в официальном браке чаще имеют задолженность.
<br>
Гипотеза подтвердилась: люди, которые никогда не состояли в официальном браке, на 2,1% чаще имеют задолженность, чем те, у кого брак в жизни был.
<br>
Так, зависимость между семейным статусом и наличием задолженности существует, однако для ее ясного обнаружения требуется иная категоризация.

Он становится менее дисциплинированным в финансовом смысле, хотя, на самом деле, еще не может себе этого позволить. Пик "антидисциплинированности" приходится на средний средний класс. Далее тенденция идет на убыль - высший средний класс и богатые реже имеют задолженность, чем средний средний класс. Во-первых, это связано с очевидным ростом доходов и, как следствие, появлением денег на "незапланированные расходы". Во-вторых, люди, переходящие в категории богатых и высшего среднего класса, вероятно, изначально более грамотны в финансовом понимании.

### 4.4 Сравнительный анализ категорий по цели кредита в отношении наличия задолженности

```python
print(new_df.groupby('purpose_categories')['debt'].agg(['sum', 'count', 'mean']).sort_values('mean'))
```

Зависимость от разных целей кредита определенно есть: чаще всего имеют задолженность те, кто брал кредит на автомобиль (9,3% случаев). Самые добропорядочные заемщики это ипотечные кредиторы: только 7,2% заемщиков имеют задолженность. Те, кто берет кредит на коммерческую недвижимость и на свадьбу имеют задолженность в 7,5% и 7,9% случаев соотвественно. Те, кто берет кредит на образование, в 9,2% случаев просрачивают выплаты.
